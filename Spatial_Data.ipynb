```python
import folium
from folium.plugins import MarkerCluster
from IPython.display import display, Image


world_map = folium.Map(location=[0, 0], zoom_start=2)


places = [
    {"name": "New York City", "coords": [40.7128, -74.0060], "image_url": "https://source.unsplash.com/400x300/?newyork"},
    {"name": "London", "coords": [51.5074, -0.1278], "image_url": "https://source.unsplash.com/400x300/?london"},
    {"name": "Tokyo", "coords": [35.6895, 139.6917], "image_url": "https://source.unsplash.com/400x300/?tokyo"},
    {"name": "Sydney", "coords": [-33.85903383272798, 151.20364678216654], "image_url": "https://source.unsplash.com/400x300/?sydney"},
    {"name": "Taj Mahal", "coords": [27.17540725072536, 78.04222802682138], "image_url": "https://source.unsplash.com/400x300/?tajmahal"},
    {"name": "Sheikh Zayed Grand Mosque", "coords": [24.412999412921067, 54.47505047393601], "image_url": "https://source.unsplash.com/400x300/?Sheikh Zayed Grand Mosque"},

]


for place in places:
    popup_content = f"<b>{place['name']}</b><br><img src='{place['image_url']}' width='200px'>"
    folium.Marker(location=place["coords"], popup=folium.Popup(html=popup_content, max_width='auto')).add_to(world_map)

display(world_map)

```
